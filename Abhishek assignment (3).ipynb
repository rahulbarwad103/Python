```python
# Question 15: Given a dictionary my_dict = {'a': 10, 'b': 20, 'c': 30}, how can you sort it by keys? Write a Python program to achieve this.


my_dict = {'a': 10, 'b': 20, 'c': 30}

sorted_dict = {key:my_dict[key]for key in sorted(my_dict.keys())}


print(sorted_dict)


```

```python
# Question 19: Given a dictionary my_dict = {'apple': 3, 'banana': 5}, how can you get the key associated with the maximum value? Write a Python program to achieve this.

my_dict = {'apple': 3, 'banana': 5}

max_key = max(my_dict, key=my_dict.get)

print("The key associated with the maximum value is:", max_key)


```
